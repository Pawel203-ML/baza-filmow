2. Pogrupuj tabelę w taki sposób, aby otrzymać średni przychód (revenue) oraz średni budżet (budget) w danym roku dla filmów opublikowanych od 2010 (włącznie) do 2016 roku (włącznie). Następnie na tej podstawie stwórz wykres, w którym średnie przychody są wykresem kolumnowym, a średni budżet wykresem liniowym na tych samych osiach. Sformatuj odpowiednio oś X oraz oś Y. Dodaj tytuł wykresu, oraz legendę, która znajduje się w prawym górnym rogu płótna, lecz poza obszarem osi. Przykład wykresu widoczny poniżej:


```python
df2[df2['release_date'].isnull()]
```

```python
df2_copy =  df2.copy()
df2_copy['release_date'].dropna()
df2_copy['release_date'] = pd.to_datetime(df2_copy['release_date']).dt.year.astype('Int64')
df2_copy = df2_copy[(df2_copy['release_date'] >= 2010) & (df2_copy['release_date'] <= 2016)]
df2_copy = df2_copy[['original_title','release_date','revenue','budget']]
df2_copy
```

liczenie sredniego budzetu i przychodu na kazdy rok

```python
average_revenue = df2_copy[['release_date', 'revenue']].groupby('release_date').mean()
average_revenue
```

```python
average_budget = df2_copy[['release_date', 'budget']].groupby('release_date').mean()
average_budget
```

```python
def milion(x, pos):
    return '{:2.1f}M'.format(x * 1e-6 )

fig = plt.figure()
axis = fig.add_axes([0,0,1,1])

axis.bar(average_revenue.index, average_revenue['revenue'], label = 'revenue')
axis.plot(average_budget.index, average_budget['budget'], linewidth = 4, color = 'red', label = 'budget')

formatter = plt.FuncFormatter(milion)
axis.yaxis.set_major_formatter(formatter)

axis.legend(loc = (1.05, 0.9))
plt.title('Porownanie budzetu i przychodu rocznego')
plt.show()
```

3. Baza filmów zawiera kolumnę z id gatunku (genre_id). Na tej podstawie połącz ze sobą bazę filmów z bazą gatunków, tak aby w bazie filmów można było odczytać nazwę gatunku filmu.

```python
df1_copy = df1.copy()
df2_copy = df2.copy()
combined_df1_df2 = pd.merge(df2_copy, df1_copy, left_on = 'genre_id', right_on = 'num', how = 'outer')
combined_df1_df2
```

Sprawdzenie czy wartosci brakujace stanowia taki sam procent co w orginalenej tabeli

```python
combined_df1_df2['genres'].isnull().mean()
```

5. Filmy, którego gatunku trwają średnio najdłużej (runtime)?

```python
average_runtime = combined_df1_df2[['runtime', 'genres']].groupby('genres').mean()
average_runtime = average_runtime.sort_values(by = 'runtime', ascending = False)
average_runtime[:1]
```
